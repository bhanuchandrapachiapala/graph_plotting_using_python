```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
n=np.linspace(1,100)
plt.plot(n,7*n*n+6*n+5, label="O(n^2)")
plt.plot(n,7*n+5, label="O(n)")
plt.legend(loc='upper left')
plt.show()
```

```python
n=np.linspace(1,7)
plt.plot(n,n*n, label="O(n^2)")
plt.plot(n,n, label="O(n)")
plt.plot(n,n*n*n, label="O(n^3)")
plt.legend(loc='upper left')
plt.show()
```

```python
x=np.linspace(1,100)
plt.plot(n,(7*n*n+6*n+5)/(n*n), label="1")
plt.plot(n,(n*n)/n, label="2")
plt.legend(loc='upper right')
plt.show()
```

```python
n=np.linspace(1,10)
plt.plot(n,n**2, label="O(n^2)")
plt.plot(n,2**n, label="O(2^n)")
plt.legend(loc='upper left')
plt.show()
```

```python
x=np.linspace(1,100)
plt.plot(n,(n*n)/n, label="2")
plt.plot(n,(7*n*n+6*n+5)/(n*n), label="1")
plt.legend(loc='best')
plt.show()
```

```python
n=np.linspace(1,10)
plt.plot(n,n**2, label="O(n^2)")
plt.plot(n,2**n, label="O(2^n)")
plt.legend(loc='center')
plt.show()
```

```python
n=np.linspace(1,20)
plt.plot(n,n, label="O(n)")
plt.plot(n,np.log(n), label="O(log(n))")
plt.legend(loc='right')
plt.show()
```

```python
n=np.linspace(1,10)
plt.plot(n,n**.5, label="n^5")
plt.plot(n,np.log(n)**3, label="log(n)^3")
plt.legend(loc='best')
plt.show()
```

```python
n=np.linspace(1,10**6)
plt.plot(n,n**.5, label="n^.5")
plt.plot(n,np.log(n)**3, label="log(n)^3")
plt.legend(loc='best')
plt.show()
```
